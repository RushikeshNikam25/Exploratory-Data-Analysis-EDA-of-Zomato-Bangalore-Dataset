```python
#Reading Column Names
df.columns
```

```python
#Changing the column names
df = df.rename(columns={'approx_cost(for two people)':'cost','listed_in(type)':'type',
                                  'listed_in(city)':'city'})
df.columns
```

```python
df.head()
```

```python
df.cost.unique()
```

```python
print(type(df['cost'])) # = df['cost'].astype(float)
```

```python
#Some Transformations
df['cost'] = df['cost'].astype(str) #Changing the cost to string
df['cost'] = df['cost'].apply(lambda x: x.replace(',','')) #Using lambda function to replace ',' from cost
df['cost'] = df['cost'].astype(float) # Changing the cost to Float
df.info() # looking at the dataset information after transformation
```

```python
print(type(df['cost']))
```

```python
df['rate'].unique()
```

```python
import pandas as pd
import numpy as np

# Assuming 'df' is the DataFrame containing Zomato Bangalore dataset

# Removing '/5' from Rates
df = df.loc[df.rate != 'NEW']  # Removed rows which were having "NEW"
df = df.loc[df.rate != '-'].reset_index(drop=True)  # Removed rows which were having "-"
remove_slash = lambda x: x.replace('/5', '') if type(x) == str else x  # Function to remove '/5'
df.rate = df.rate.apply(remove_slash).str.strip().astype(float)  # Apply function to column
df['rate'].head()  # Displaying the dataset after transformation
```
